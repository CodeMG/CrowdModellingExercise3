### Task 2/5: Common bifurcations in nonlinear systems

\begin{align}
\dot{x} & = \alpha - x^2
\end{align}

At $\alpha = 0$ there is a fold bifurcation as it is associated with the appearance of an eigenvalue $\lambda_1 = 0$

#### Bifurcation diagram for $\alpha$ in $(-1, 1)$

$f(x, \alpha) = 0 => x = \pm\sqrt\alpha$

```python
import numpy as np
```

```python
x = lambda alpha: [np.sqrt(alpha), -np.sqrt(alpha)] if alpha >= 0 else None
x_dot = lambda x, alpha: alpha - x*x
```

```python
alpha = np.arange(-1,1,0.01)
```

```python
x_val = [x(a) for a in alpha]
```

```python
ARROW_LENGTH = 0.1
def make_arrow(alpha, x, x_dot):
    return (alpha, x, 0, np.sign(x_dot(x, alpha)) * ARROW_LENGTH)

arrows = []
for i in range(0, len(alpha), 10):
    if x_val[i]:
        [arrows.append(make_arrow(alpha[i], xx - 2 * ARROW_LENGTH, x_dot)) for xx in x_val[i]]
        [arrows.append(make_arrow(alpha[i], xx + 2 * ARROW_LENGTH, x_dot)) for xx in x_val[i]]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(alpha, [xx[0] if xx else None for xx in x_val])
ax.plot(alpha, [xx[1] if xx else None for xx in x_val])

for arrow in arrows:
    ax.arrow(arrow[0], arrow[1], arrow[2], arrow[3], head_width=0.01, head_length=0.05)

plt.xlim([-1, 1])
plt.xlabel(r'$\alpha$')
plt.ylabel('x')
plt.show()
```

#### Bifurcation diagram for $\alpha$ in $(2, 4)$

$f(x, \alpha) = 0 => x = \pm\sqrt{\frac{\alpha-3}{2}}$

```python
x = lambda alpha: [np.sqrt(0.5*(alpha-3)), -np.sqrt(0.5*(alpha-3))] if alpha >= 3 else None
x_dot = lambda x, alpha: alpha - 2*x*x - 3 
```

```python
alpha = np.arange(2,4,0.001)
```

```python
x_val = [x(a) for a in alpha]
```

```python
arrows = []
for i in range(0, len(alpha), 100):
    if x_val[i]:
        [arrows.append(make_arrow(alpha[i], xx - 2 * ARROW_LENGTH, x_dot)) for xx in x_val[i]]
        [arrows.append(make_arrow(alpha[i], xx + 2 * ARROW_LENGTH, x_dot)) for xx in x_val[i]]
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(alpha, [xx[0] if xx else None for xx in x_val])
ax.plot(alpha, [xx[1] if xx else None for xx in x_val])

for arrow in arrows:
    ax.arrow(arrow[0], arrow[1], arrow[2], arrow[3], head_width=0.01, head_length=0.05)

plt.xlim([2, 4])
plt.xlabel(r'$\alpha$')
plt.ylabel('x')
plt.show()
```

At $\alpha=1$ the two dynamical systems take the following forms:
\begin{align}
\dot{x} & = 1 - x^2
\end{align}

\begin{align}
\dot{x} & = -2 - 2x^2
\end{align}

At this value of $\alpha$ they are not topologically equivalent, as their phase portraits cannot be obtained from one another by a continous transformation.

At $\alpha=1$ the two dynamical systems take the following forms:
\begin{align}
\dot{x} & = -1 - x^2
\end{align}

\begin{align}
\dot{x} & = -4 - 2x^2
\end{align}

At this value of $\alpha$ they are topologically equivalent, as there exists a homeomorphism $h = \sqrt{2}x$, which maps orbits of the first system onto orbits of the second system.
